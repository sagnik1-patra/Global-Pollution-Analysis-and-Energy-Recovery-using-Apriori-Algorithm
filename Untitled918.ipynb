```python
import os
import json
import yaml
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    roc_curve,
    auc
)

from sklearn.ensemble import RandomForestClassifier

# Install if needed:
# pip install mlxtend tensorflow pyyaml seaborn

import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Flatten, Dropout, MaxPooling2D
from tensorflow.keras.utils import to_categorical


# ============================================================
# Paths
# ============================================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Global Pollution Analysis and Energy Recovery using Apriori Algorithm\Global_Pollution_Analysis.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Global Pollution Analysis and Energy Recovery using Apriori Algorithm"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "apriori_"


# ============================================================
# Utility Save Function
# ============================================================

def save_fig(name):
    path = os.path.join(OUTPUT_DIR, PREFIX + name)
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", path)


def save_csv(df, name):
    path = os.path.join(OUTPUT_DIR, PREFIX + name)
    df.to_csv(path, index=False)
    print("Saved:", path)


def save_json(data, name):
    path = os.path.join(OUTPUT_DIR, PREFIX + name)
    with open(path, "w") as f:
        json.dump(data, f, indent=4)
    print("Saved:", path)


def save_yaml(data, name):
    path = os.path.join(OUTPUT_DIR, PREFIX + name)
    with open(path, "w") as f:
        yaml.dump(data, f)
    print("Saved:", path)


# ============================================================
# Phase 1: Data Loading
# ============================================================

df = pd.read_csv(DATA_PATH)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())

save_csv(df, "raw_dataset.csv")


# ============================================================
# Data Cleaning
# ============================================================

df.columns = df.columns.str.strip()

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].mean())

df = df.drop_duplicates()

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

save_csv(df, "cleaned_dataset.csv")


# ============================================================
# Feature Engineering
# ============================================================

air_col = "Air_Pollution_Index"
water_col = "Water_Pollution_Index"
soil_col = "Soil_Pollution_Index"
energy_col = "Energy_Recovered (in GWh)"
population_col = "Population (in millions)"
consumption_col = "Energy_Consumption_Per_Capita (in MWh)"

# Energy recovered per capita
df["Energy_Recovered_Per_Capita"] = df[energy_col] / df[population_col]

# Pollution average
df["Average_Pollution_Index"] = df[[air_col, water_col, soil_col]].mean(axis=1)

# Energy efficiency score
df["Energy_Efficiency_Score"] = df[energy_col] / (df[consumption_col] + 1)

# Severity categories
def severity_category(value):
    if value < 100:
        return "Low"
    elif value < 200:
        return "Medium"
    else:
        return "High"

df["Air_Severity"] = df[air_col].apply(severity_category)
df["Water_Severity"] = df[water_col].apply(severity_category)
df["Soil_Severity"] = df[soil_col].apply(severity_category)
df["Overall_Severity"] = df["Average_Pollution_Index"].apply(severity_category)

# Energy recovered category
df["Energy_Recovery_Level"] = pd.qcut(
    df[energy_col],
    q=3,
    labels=["Low_Recovery", "Medium_Recovery", "High_Recovery"]
)

# Energy consumption category
df["Energy_Consumption_Level"] = pd.qcut(
    df[consumption_col],
    q=3,
    labels=["Low_Consumption", "Medium_Consumption", "High_Consumption"]
)

save_csv(df, "feature_engineered_dataset.csv")


# ============================================================
# Normalize Pollution Indices
# ============================================================

scaler = MinMaxScaler()

pollution_cols = [
    air_col,
    water_col,
    soil_col,
    "Average_Pollution_Index",
    "Energy_Recovered_Per_Capita",
    "Energy_Efficiency_Score"
]

df_scaled = df.copy()
df_scaled[pollution_cols] = scaler.fit_transform(df_scaled[pollution_cols])

save_csv(df_scaled, "normalized_dataset.csv")


# ============================================================
# Encode Categorical Features
# ============================================================

df_encoded = df_scaled.copy()
label_encoders = {}

for col in df_encoded.select_dtypes(include=["object", "category"]).columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
    label_encoders[col] = list(le.classes_)

save_csv(df_encoded, "encoded_dataset.csv")
save_json(label_encoders, "label_encoders.json")


# ============================================================
# Pollution Trend Analysis
# ============================================================

trend_df = df.groupby("Year")[
    [air_col, water_col, soil_col, energy_col]
].mean().reset_index()

save_csv(trend_df, "pollution_trend_analysis.csv")

plt.figure(figsize=(10, 6))
plt.plot(trend_df["Year"], trend_df[air_col], marker="o", label="Air Pollution")
plt.plot(trend_df["Year"], trend_df[water_col], marker="o", label="Water Pollution")
plt.plot(trend_df["Year"], trend_df[soil_col], marker="o", label="Soil Pollution")
plt.xlabel("Year")
plt.ylabel("Average Pollution Index")
plt.title("Yearly Pollution Trends")
plt.legend()
plt.grid(True)
save_fig("pollution_trend_graph.png")

plt.figure(figsize=(8, 6))
sns.heatmap(df_encoded.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
save_fig("correlation_heatmap.png")


# ============================================================
# Phase 2: Apriori Algorithm
# ============================================================

basket_df = df[
    [
        "Country",
        "Air_Severity",
        "Water_Severity",
        "Soil_Severity",
        "Overall_Severity",
        "Energy_Recovery_Level",
        "Energy_Consumption_Level"
    ]
].copy()

basket_df["Country"] = "Country_" + basket_df["Country"].astype(str)
basket_df["Air_Severity"] = "Air_" + basket_df["Air_Severity"].astype(str)
basket_df["Water_Severity"] = "Water_" + basket_df["Water_Severity"].astype(str)
basket_df["Soil_Severity"] = "Soil_" + basket_df["Soil_Severity"].astype(str)
basket_df["Overall_Severity"] = "Overall_" + basket_df["Overall_Severity"].astype(str)
basket_df["Energy_Recovery_Level"] = basket_df["Energy_Recovery_Level"].astype(str)
basket_df["Energy_Consumption_Level"] = basket_df["Energy_Consumption_Level"].astype(str)

transactions = pd.get_dummies(basket_df)

save_csv(transactions, "apriori_transaction_matrix.csv")

frequent_itemsets = apriori(
    transactions,
    min_support=0.05,
    use_colnames=True
)

frequent_itemsets["itemsets"] = frequent_itemsets["itemsets"].apply(lambda x: ", ".join(list(x)))

save_csv(frequent_itemsets, "frequent_itemsets.csv")

rules = association_rules(
    apriori(transactions, min_support=0.05, use_colnames=True),
    metric="confidence",
    min_threshold=0.4
)

rules["antecedents"] = rules["antecedents"].apply(lambda x: ", ".join(list(x)))
rules["consequents"] = rules["consequents"].apply(lambda x: ", ".join(list(x)))

rules = rules.sort_values(by=["lift", "confidence"], ascending=False)

save_csv(rules, "association_rules.csv")

print("\nTop Association Rules:")
print(rules.head(10))


# ============================================================
# Frequent Itemset Visualization
# ============================================================

top_itemsets = frequent_itemsets.sort_values("support", ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.barh(top_itemsets["itemsets"], top_itemsets["support"])
plt.xlabel("Support")
plt.ylabel("Itemsets")
plt.title("Top Frequent Itemsets")
plt.gca().invert_yaxis()
save_fig("frequent_itemsets_graph.png")


# ============================================================
# Association Rule Visualization
# ============================================================

top_rules = rules.head(20)

plt.figure(figsize=(10, 6))
plt.scatter(top_rules["support"], top_rules["confidence"], s=top_rules["lift"] * 60)
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.title("Association Rules: Support vs Confidence")
save_fig("association_rules_scatter_graph.png")

plt.figure(figsize=(10, 6))
plt.barh(top_rules["antecedents"] + " → " + top_rules["consequents"], top_rules["lift"])
plt.xlabel("Lift")
plt.ylabel("Rules")
plt.title("Top Association Rules by Lift")
plt.gca().invert_yaxis()
save_fig("association_rules_lift_graph.png")


# ============================================================
# Phase 3: Apriori Validation
# ============================================================

train_data, test_data = train_test_split(transactions, test_size=0.2, random_state=42)

train_itemsets = apriori(train_data, min_support=0.05, use_colnames=True)
train_rules = association_rules(
    train_itemsets,
    metric="confidence",
    min_threshold=0.4
)

validation_results = []

for _, rule in train_rules.iterrows():
    antecedent = list(rule["antecedents"])
    consequent = list(rule["consequents"])

    antecedent_match = test_data[antecedent].all(axis=1)

    if antecedent_match.sum() > 0:
        consequent_match = test_data.loc[antecedent_match, consequent].all(axis=1)
        rule_accuracy = consequent_match.mean()
    else:
        rule_accuracy = 0

    validation_results.append({
        "antecedents": ", ".join(antecedent),
        "consequents": ", ".join(consequent),
        "support": rule["support"],
        "confidence": rule["confidence"],
        "lift": rule["lift"],
        "test_rule_accuracy": rule_accuracy
    })

validation_df = pd.DataFrame(validation_results)
validation_df = validation_df.sort_values("test_rule_accuracy", ascending=False)

save_csv(validation_df, "apriori_validation_results.csv")

print("\nApriori Validation Results:")
print(validation_df.head())


# ============================================================
# K-Fold Validation for Apriori Rules
# ============================================================

kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_results = []

for fold, (train_idx, test_idx) in enumerate(kf.split(transactions), start=1):
    train_fold = transactions.iloc[train_idx]
    test_fold = transactions.iloc[test_idx]

    fold_itemsets = apriori(train_fold, min_support=0.05, use_colnames=True)

    if len(fold_itemsets) == 0:
        continue

    fold_rules = association_rules(
        fold_itemsets,
        metric="confidence",
        min_threshold=0.4
    )

    if len(fold_rules) == 0:
        continue

    accuracies = []

    for _, rule in fold_rules.iterrows():
        antecedent = list(rule["antecedents"])
        consequent = list(rule["consequents"])

        antecedent_match = test_fold[antecedent].all(axis=1)

        if antecedent_match.sum() > 0:
            consequent_match = test_fold.loc[antecedent_match, consequent].all(axis=1)
            accuracies.append(consequent_match.mean())

    cv_results.append({
        "Fold": fold,
        "Number_of_Rules": len(fold_rules),
        "Average_Confidence": fold_rules["confidence"].mean(),
        "Average_Lift": fold_rules["lift"].mean(),
        "Average_Test_Rule_Accuracy": np.mean(accuracies) if accuracies else 0
    })

cv_df = pd.DataFrame(cv_results)
save_csv(cv_df, "apriori_cross_validation_results.csv")

plt.figure(figsize=(8, 5))
plt.plot(cv_df["Fold"], cv_df["Average_Test_Rule_Accuracy"], marker="o")
plt.xlabel("Fold")
plt.ylabel("Average Rule Accuracy")
plt.title("Apriori Cross Validation Accuracy")
plt.grid(True)
save_fig("apriori_cross_validation_graph.png")


# ============================================================
# Phase 4: CNN Classification Model
# Target: Overall Pollution Severity
# ============================================================

cnn_df = df_encoded.copy()

target_col = "Overall_Severity"

X = cnn_df.drop(columns=[target_col])
y = cnn_df[target_col]

num_classes = len(np.unique(y))

X_scaled = MinMaxScaler().fit_transform(X)

# Convert tabular data into small image format
feature_count = X_scaled.shape[1]
image_size = int(np.ceil(np.sqrt(feature_count)))
required_features = image_size * image_size

padding = required_features - feature_count
X_padded = np.pad(X_scaled, ((0, 0), (0, padding)), mode="constant")

X_images = X_padded.reshape(-1, image_size, image_size, 1)
y_cat = to_categorical(y, num_classes=num_classes)

X_train, X_test, y_train, y_test = train_test_split(
    X_images,
    y_cat,
    test_size=0.2,
    random_state=42,
    stratify=y
)

cnn_model = Sequential([
    Conv2D(32, (2, 2), activation="relu", input_shape=(image_size, image_size, 1)),
    MaxPooling2D((1, 1)),
    Dropout(0.2),

    Conv2D(64, (2, 2), activation="relu"),
    Flatten(),

    Dense(64, activation="relu"),
    Dropout(0.3),

    Dense(num_classes, activation="softmax")
])

cnn_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = cnn_model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=30,
    batch_size=16,
    verbose=1
)

cnn_model_path = os.path.join(OUTPUT_DIR, PREFIX + "cnn_pollution_model.h5")
cnn_model.save(cnn_model_path)
print("Saved:", cnn_model_path)


# ============================================================
# CNN Evaluation
# ============================================================

loss, cnn_accuracy = cnn_model.evaluate(X_test, y_test, verbose=0)

y_pred_prob = cnn_model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)
y_true = np.argmax(y_test, axis=1)

cnn_report = classification_report(y_true, y_pred, output_dict=True)
cnn_report_df = pd.DataFrame(cnn_report).transpose()

save_csv(cnn_report_df, "cnn_classification_report.csv")

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("CNN Confusion Matrix")
save_fig("cnn_confusion_matrix.png")

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("CNN Accuracy Graph")
plt.legend()
plt.grid(True)
save_fig("cnn_accuracy_graph.png")

plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("CNN Loss Graph")
plt.legend()
plt.grid(True)
save_fig("cnn_loss_graph.png")


# ============================================================
# ROC Curve
# ============================================================

plt.figure(figsize=(8, 6))

for i in range(num_classes):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred_prob[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"Class {i} AUC = {roc_auc:.2f}")

plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("CNN ROC Curve")
plt.legend()
plt.grid(True)
save_fig("cnn_roc_curve.png")


# ============================================================
# Random Forest Baseline for Comparison
# ============================================================

X_ml = df_encoded.drop(columns=[target_col])
y_ml = df_encoded[target_col]

X_train_ml, X_test_ml, y_train_ml, y_test_ml = train_test_split(
    X_ml,
    y_ml,
    test_size=0.2,
    random_state=42,
    stratify=y_ml
)

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_ml, y_train_ml)

rf_pred = rf_model.predict(X_test_ml)
rf_accuracy = accuracy_score(y_test_ml, rf_pred)

comparison_df = pd.DataFrame({
    "Model": ["CNN", "Random Forest", "Apriori Rule Mining"],
    "Accuracy_or_Score": [
        cnn_accuracy,
        rf_accuracy,
        validation_df["test_rule_accuracy"].mean() if len(validation_df) > 0 else 0
    ]
})

save_csv(comparison_df, "model_comparison_results.csv")

plt.figure(figsize=(8, 5))
plt.bar(comparison_df["Model"], comparison_df["Accuracy_or_Score"])
plt.xlabel("Model")
plt.ylabel("Accuracy / Rule Score")
plt.title("Model Comparison Graph")
save_fig("model_comparison_graph.png")


# ============================================================
# Prediction CSV
# ============================================================

prediction_df = pd.DataFrame({
    "Actual_Class": y_true,
    "Predicted_Class": y_pred
})

for i in range(num_classes):
    prediction_df[f"Class_{i}_Probability"] = y_pred_prob[:, i]

save_csv(prediction_df, "cnn_prediction_results.csv")

plt.figure(figsize=(10, 5))
plt.plot(prediction_df["Actual_Class"].values, label="Actual")
plt.plot(prediction_df["Predicted_Class"].values, label="Predicted")
plt.xlabel("Sample")
plt.ylabel("Class")
plt.title("CNN Actual vs Predicted Pollution Severity")
plt.legend()
plt.grid(True)
save_fig("cnn_actual_vs_predicted_graph.png")


# ============================================================
# Save Configuration
# ============================================================

config = {
    "project_name": "Global Pollution Analysis and Energy Recovery using Apriori Algorithm",
    "dataset": DATA_PATH,
    "output_directory": OUTPUT_DIR,
    "target_column": target_col,
    "apriori_min_support": 0.05,
    "apriori_min_confidence": 0.4,
    "cnn_epochs": 30,
    "cnn_batch_size": 16,
    "cnn_accuracy": float(cnn_accuracy),
    "random_forest_accuracy": float(rf_accuracy),
    "average_apriori_rule_accuracy": float(validation_df["test_rule_accuracy"].mean()) if len(validation_df) > 0 else 0
}

save_json(config, "project_summary.json")
save_yaml(config, "project_config.yaml")


# ============================================================
# Final Report
# ============================================================

top_rule_text = rules.head(10).to_string(index=False) if len(rules) > 0 else "No strong rules found."

report = f"""
# Global Pollution Analysis and Energy Recovery using Apriori Algorithm

## Objective
The objective of this project is to analyze global pollution data and discover hidden associations between pollution levels, energy consumption, and energy recovery using the Apriori Algorithm.

## Dataset
Dataset used:
{DATA_PATH}

Total records: {df.shape[0]}
Total features: {df.shape[1]}

## Phase 1: Data Preprocessing
The dataset was loaded, cleaned, and processed. Missing values were handled using mean imputation for numerical columns and mode imputation for categorical columns.

## Feature Engineering
The following features were created:
- Energy_Recovered_Per_Capita
- Average_Pollution_Index
- Energy_Efficiency_Score
- Air_Severity
- Water_Severity
- Soil_Severity
- Overall_Severity
- Energy_Recovery_Level
- Energy_Consumption_Level

## Phase 2: Apriori Algorithm
The Apriori Algorithm was applied to identify association rules among pollution severity, energy recovery, and energy consumption.

Minimum Support: 0.05  
Minimum Confidence: 0.4

## Top Association Rules
{top_rule_text}

## Phase 3: Validation
Train-test validation and 5-fold cross-validation were used to validate Apriori rules.

Average Apriori Rule Accuracy:
{config["average_apriori_rule_accuracy"]}

## CNN Model
A CNN model was trained by converting tabular pollution features into image-like matrices.

CNN Accuracy:
{cnn_accuracy}

## Random Forest Baseline
Random Forest Accuracy:
{rf_accuracy}

## Key Insights
- Apriori rules help identify relationships between pollution severity and energy recovery.
- High pollution categories can be associated with different recovery and consumption levels.
- CNN provides classification capability for pollution severity prediction.
- Rule mining provides interpretability, while CNN provides predictive performance.

## Recommendations
- Countries with high pollution and low recovery should improve renewable energy adoption.
- Energy recovery strategies should be prioritized in high air and water pollution regions.
- Association rules can help policymakers identify recurring pollution-energy patterns.
- Machine learning models can support future pollution severity prediction.

## Generated Outputs
- Cleaned dataset
- Normalized dataset
- Encoded dataset
- Frequent itemsets CSV
- Association rules CSV
- Apriori validation CSV
- Cross-validation CSV
- CNN model H5 file
- CNN prediction CSV
- Confusion matrix
- ROC curve
- Correlation heatmap
- Frequent itemset graph
- Association rule graphs
- Model comparison graph
"""

report_path = os.path.join(OUTPUT_DIR, PREFIX + "final_report.md")
with open(report_path, "w", encoding="utf-8") as f:
    f.write(report)

print("Saved:", report_path)


# ============================================================
# Final Output Summary
# ============================================================

print("\nProject Completed Successfully!")
print("All outputs saved in:")
print(OUTPUT_DIR)
```
