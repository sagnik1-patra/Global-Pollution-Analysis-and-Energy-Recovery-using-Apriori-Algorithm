````python
# ============================================================
# Global Pollution Analysis and Energy Recovery using Apriori
# Fixed Full Code - No README, No ZIP
# Author: Sagnik Patra
# ============================================================

import os
import json
import yaml
import warnings
warnings.filterwarnings("ignore")

import nbformat
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.frequent_patterns import apriori, association_rules

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    roc_curve,
    auc
)
from sklearn.ensemble import RandomForestClassifier

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical


# ============================================================
# Paths
# ============================================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Global Pollution Analysis and Energy Recovery using Apriori Algorithm\Global_Pollution_Analysis.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Global Pollution Analysis and Energy Recovery using Apriori Algorithm"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "apriori_"


# ============================================================
# Save Helpers
# ============================================================

def save_csv(dataframe, filename):
    path = os.path.join(OUTPUT_DIR, PREFIX + filename)
    dataframe.to_csv(path, index=False)
    print("Saved:", path)


def save_fig(filename):
    path = os.path.join(OUTPUT_DIR, PREFIX + filename)
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", path)


def save_json(data, filename):
    path = os.path.join(OUTPUT_DIR, PREFIX + filename)
    with open(path, "w", encoding="utf-8") as file:
        json.dump(data, file, indent=4)
    print("Saved:", path)


def save_yaml(data, filename):
    path = os.path.join(OUTPUT_DIR, PREFIX + filename)
    with open(path, "w", encoding="utf-8") as file:
        yaml.dump(data, file)
    print("Saved:", path)


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())


# ============================================================
# Auto Column Finder
# ============================================================

def find_col(possible_names):
    for name in possible_names:
        if name in df.columns:
            return name

    lower_map = {col.lower().strip(): col for col in df.columns}

    for name in possible_names:
        key = name.lower().strip()
        if key in lower_map:
            return lower_map[key]

    raise ValueError("Column not found. Tried: " + str(possible_names))


country_col = find_col(["Country", "country"])
year_col = find_col(["Year", "year"])

air_col = find_col([
    "Air_Pollution_Index",
    "Air Pollution Index",
    "Air_Pollution",
    "Air Pollution"
])

water_col = find_col([
    "Water_Pollution_Index",
    "Water Pollution Index",
    "Water_Pollution",
    "Water Pollution"
])

soil_col = find_col([
    "Soil_Pollution_Index",
    "Soil Pollution Index",
    "Soil_Pollution",
    "Soil Pollution"
])

energy_col = find_col([
    "Energy_Recovered (in GWh)",
    "Energy Recovered",
    "Energy_Recovered",
    "Energy Recovery",
    "Energy_Recovery"
])

population_col = find_col([
    "Population (in millions)",
    "Population",
    "population"
])

consumption_col = find_col([
    "Energy_Consumption_Per_Capita (in MWh)",
    "Energy Consumption Per Capita",
    "Energy_Consumption_Per_Capita",
    "Energy Consumption",
    "Energy_Consumption"
])


# ============================================================
# Data Cleaning
# ============================================================

for col in df.columns:
    if df[col].dtype == "object":
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].mean())

df.drop_duplicates(inplace=True)


# ============================================================
# Feature Engineering
# ============================================================

df["Energy_Recovered_Per_Capita"] = df[energy_col] / (df[population_col] + 1)
df["Energy_Consumption_Efficiency"] = df[energy_col] / (df[consumption_col] + 1)
df["Average_Pollution_Index"] = df[[air_col, water_col, soil_col]].mean(axis=1)


def severity_category(value):
    if value < 100:
        return "Low"
    elif value < 200:
        return "Medium"
    else:
        return "High"


df["Air_Severity"] = df[air_col].apply(severity_category)
df["Water_Severity"] = df[water_col].apply(severity_category)
df["Soil_Severity"] = df[soil_col].apply(severity_category)
df["Overall_Severity"] = df["Average_Pollution_Index"].apply(severity_category)

df["Energy_Recovery_Level"] = pd.qcut(
    df[energy_col],
    q=3,
    labels=["Low_Recovery", "Medium_Recovery", "High_Recovery"],
    duplicates="drop"
)

df["Energy_Consumption_Level"] = pd.qcut(
    df[consumption_col],
    q=3,
    labels=["Low_Consumption", "Medium_Consumption", "High_Consumption"],
    duplicates="drop"
)

save_csv(df, "processed_features.csv")


# ============================================================
# Pollution Trend Analysis
# ============================================================

trend_df = df.groupby(year_col)[
    [air_col, water_col, soil_col, energy_col]
].mean().reset_index()

save_csv(trend_df, "pollution_trend_summary.csv")

plt.figure(figsize=(10, 6))
plt.plot(trend_df[year_col], trend_df[air_col], marker="o", label="Air Pollution")
plt.plot(trend_df[year_col], trend_df[water_col], marker="o", label="Water Pollution")
plt.plot(trend_df[year_col], trend_df[soil_col], marker="o", label="Soil Pollution")
plt.xlabel("Year")
plt.ylabel("Average Pollution Index")
plt.title("Yearly Pollution Trend Analysis")
plt.legend()
plt.grid(True)
save_fig("pollution_trend_graph.png")


# ============================================================
# Encoding and Normalization
# ============================================================

encoded_df = df.copy()
label_encoders = {}

for col in encoded_df.select_dtypes(include=["object", "category"]).columns:
    le = LabelEncoder()
    encoded_df[col] = le.fit_transform(encoded_df[col].astype(str))
    label_encoders[col] = list(le.classes_)

pollution_scale_cols = [
    air_col,
    water_col,
    soil_col,
    "Average_Pollution_Index",
    "Energy_Recovered_Per_Capita",
    "Energy_Consumption_Efficiency"
]

scaler = MinMaxScaler()
encoded_df[pollution_scale_cols] = scaler.fit_transform(encoded_df[pollution_scale_cols])

save_csv(encoded_df, "encoded_normalized_features.csv")
save_json(label_encoders, "label_encoders.json")


# ============================================================
# Correlation Heatmap
# ============================================================

plt.figure(figsize=(12, 8))
sns.heatmap(encoded_df.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap")
save_fig("correlation_heatmap.png")


# ============================================================
# Apriori Transaction Matrix
# ============================================================

basket_df = df[
    [
        "Air_Severity",
        "Water_Severity",
        "Soil_Severity",
        "Overall_Severity",
        "Energy_Recovery_Level",
        "Energy_Consumption_Level"
    ]
].copy()

basket_df["Air_Severity"] = "Air_" + basket_df["Air_Severity"].astype(str)
basket_df["Water_Severity"] = "Water_" + basket_df["Water_Severity"].astype(str)
basket_df["Soil_Severity"] = "Soil_" + basket_df["Soil_Severity"].astype(str)
basket_df["Overall_Severity"] = "Overall_" + basket_df["Overall_Severity"].astype(str)
basket_df["Energy_Recovery_Level"] = "Energy_" + basket_df["Energy_Recovery_Level"].astype(str)
basket_df["Energy_Consumption_Level"] = "Consumption_" + basket_df["Energy_Consumption_Level"].astype(str)

transaction_df = pd.get_dummies(basket_df).astype(bool)

save_csv(transaction_df.astype(int), "transaction_matrix.csv")


# ============================================================
# Apriori Frequent Itemsets
# ============================================================

frequent_itemsets_raw = apriori(
    transaction_df,
    min_support=0.05,
    use_colnames=True
)

frequent_itemsets = frequent_itemsets_raw.copy()
frequent_itemsets["itemsets"] = frequent_itemsets["itemsets"].apply(
    lambda x: ", ".join(list(x))
)
frequent_itemsets = frequent_itemsets.sort_values("support", ascending=False)

save_csv(frequent_itemsets, "frequent_itemsets.csv")


# ============================================================
# Association Rules
# ============================================================

rules_raw = association_rules(
    frequent_itemsets_raw,
    metric="confidence",
    min_threshold=0.40
)

rules = rules_raw.copy()
rules["antecedents"] = rules["antecedents"].apply(lambda x: ", ".join(list(x)))
rules["consequents"] = rules["consequents"].apply(lambda x: ", ".join(list(x)))
rules = rules.sort_values(by=["lift", "confidence"], ascending=False)

save_csv(rules, "association_rules.csv")


# ============================================================
# Apriori Visualizations
# ============================================================

top_itemsets = frequent_itemsets.head(15)

plt.figure(figsize=(12, 6))
plt.barh(top_itemsets["itemsets"], top_itemsets["support"])
plt.xlabel("Support")
plt.ylabel("Frequent Itemsets")
plt.title("Top Frequent Itemsets")
plt.gca().invert_yaxis()
save_fig("frequent_itemsets_graph.png")

top_rules = rules.head(15)

plt.figure(figsize=(10, 6))
plt.scatter(
    top_rules["support"],
    top_rules["confidence"],
    s=top_rules["lift"] * 60
)
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.title("Association Rules: Support vs Confidence")
save_fig("association_rules_scatter_graph.png")

plt.figure(figsize=(12, 7))
plt.barh(
    top_rules["antecedents"] + " → " + top_rules["consequents"],
    top_rules["lift"]
)
plt.xlabel("Lift")
plt.ylabel("Rules")
plt.title("Top Association Rules by Lift")
plt.gca().invert_yaxis()
save_fig("association_rules_lift_graph.png")


# ============================================================
# Apriori Train-Test Validation
# ============================================================

train_data, test_data = train_test_split(
    transaction_df,
    test_size=0.2,
    random_state=42
)

train_itemsets = apriori(
    train_data,
    min_support=0.05,
    use_colnames=True
)

train_rules = association_rules(
    train_itemsets,
    metric="confidence",
    min_threshold=0.40
)

validation_results = []

for _, rule in train_rules.iterrows():
    antecedent = list(rule["antecedents"])
    consequent = list(rule["consequents"])

    antecedent_match = test_data[antecedent].all(axis=1)

    if antecedent_match.sum() > 0:
        consequent_match = test_data.loc[antecedent_match, consequent].all(axis=1)
        rule_accuracy = consequent_match.mean()
    else:
        rule_accuracy = 0

    validation_results.append({
        "antecedents": ", ".join(antecedent),
        "consequents": ", ".join(consequent),
        "support": rule["support"],
        "confidence": rule["confidence"],
        "lift": rule["lift"],
        "test_rule_accuracy": rule_accuracy
    })

validation_df = pd.DataFrame(validation_results)
validation_df = validation_df.sort_values("test_rule_accuracy", ascending=False)

save_csv(validation_df, "apriori_validation_results.csv")


# ============================================================
# Apriori K-Fold Cross Validation
# ============================================================

kf = KFold(n_splits=5, shuffle=True, random_state=42)
cv_results = []

for fold, (train_idx, test_idx) in enumerate(kf.split(transaction_df), start=1):
    train_fold = transaction_df.iloc[train_idx]
    test_fold = transaction_df.iloc[test_idx]

    fold_itemsets = apriori(
        train_fold,
        min_support=0.05,
        use_colnames=True
    )

    if fold_itemsets.empty:
        continue

    fold_rules = association_rules(
        fold_itemsets,
        metric="confidence",
        min_threshold=0.40
    )

    rule_accuracies = []

    for _, rule in fold_rules.iterrows():
        antecedent = list(rule["antecedents"])
        consequent = list(rule["consequents"])

        antecedent_match = test_fold[antecedent].all(axis=1)

        if antecedent_match.sum() > 0:
            consequent_match = test_fold.loc[antecedent_match, consequent].all(axis=1)
            rule_accuracies.append(consequent_match.mean())

    cv_results.append({
        "Fold": fold,
        "Number_of_Rules": len(fold_rules),
        "Average_Confidence": fold_rules["confidence"].mean(),
        "Average_Lift": fold_rules["lift"].mean(),
        "Average_Rule_Accuracy": np.mean(rule_accuracies) if rule_accuracies else 0
    })

cv_df = pd.DataFrame(cv_results)

save_csv(cv_df, "apriori_cross_validation_results.csv")

plt.figure(figsize=(8, 5))
plt.plot(cv_df["Fold"], cv_df["Average_Rule_Accuracy"], marker="o")
plt.xlabel("Fold")
plt.ylabel("Average Rule Accuracy")
plt.title("Apriori Cross Validation Rule Accuracy")
plt.grid(True)
save_fig("apriori_cross_validation_graph.png")


# ============================================================
# CNN Model Preparation
# ============================================================

target_col = "Overall_Severity"

cnn_df = encoded_df.copy()

X = cnn_df.drop(columns=[target_col])
y = cnn_df[target_col]

X_scaled = MinMaxScaler().fit_transform(X)

num_classes = len(np.unique(y))
y_cat = to_categorical(y, num_classes=num_classes)

feature_count = X_scaled.shape[1]
image_size = int(np.ceil(np.sqrt(feature_count)))
required_features = image_size * image_size

padding = required_features - feature_count
X_padded = np.pad(X_scaled, ((0, 0), (0, padding)), mode="constant")
X_images = X_padded.reshape(-1, image_size, image_size, 1)

X_train, X_test, y_train, y_test = train_test_split(
    X_images,
    y_cat,
    test_size=0.2,
    random_state=42,
    stratify=y
)


# ============================================================
# CNN Architecture
# ============================================================

cnn_model = Sequential([
    Conv2D(32, (2, 2), activation="relu", input_shape=(image_size, image_size, 1)),
    MaxPooling2D((1, 1)),
    Dropout(0.2),

    Conv2D(64, (2, 2), activation="relu"),
    Flatten(),

    Dense(64, activation="relu"),
    Dropout(0.3),

    Dense(num_classes, activation="softmax")
])

cnn_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = cnn_model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=30,
    batch_size=16,
    verbose=1
)


# ============================================================
# Save CNN Model
# ============================================================

model_path = os.path.join(OUTPUT_DIR, PREFIX + "cnn_pollution_model.h5")
cnn_model.save(model_path)
print("Saved:", model_path)


# ============================================================
# CNN Evaluation
# ============================================================

loss, cnn_accuracy = cnn_model.evaluate(X_test, y_test, verbose=0)

y_pred_prob = cnn_model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)
y_true = np.argmax(y_test, axis=1)

cnn_report_df = pd.DataFrame(
    classification_report(y_true, y_pred, output_dict=True)
).transpose()

save_csv(cnn_report_df, "cnn_classification_report.csv")

prediction_df = pd.DataFrame({
    "Actual_Class": y_true,
    "Predicted_Class": y_pred
})

for i in range(num_classes):
    prediction_df[f"Class_{i}_Probability"] = y_pred_prob[:, i]

save_csv(prediction_df, "cnn_prediction_results.csv")


# ============================================================
# CNN Graphs
# ============================================================

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("CNN Confusion Matrix")
save_fig("cnn_confusion_matrix.png")

plt.figure(figsize=(8, 5))
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("CNN Accuracy Graph")
plt.legend()
plt.grid(True)
save_fig("cnn_accuracy_graph.png")

plt.figure(figsize=(8, 5))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("CNN Loss Graph")
plt.legend()
plt.grid(True)
save_fig("cnn_loss_graph.png")

plt.figure(figsize=(8, 6))
for i in range(num_classes):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_pred_prob[:, i])
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"Class {i} AUC = {roc_auc:.2f}")

plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("CNN ROC Curve")
plt.legend()
plt.grid(True)
save_fig("cnn_roc_curve.png")

plt.figure(figsize=(10, 5))
plt.plot(prediction_df["Actual_Class"].values, label="Actual")
plt.plot(prediction_df["Predicted_Class"].values, label="Predicted")
plt.xlabel("Sample")
plt.ylabel("Class")
plt.title("CNN Actual vs Predicted Pollution Severity")
plt.legend()
plt.grid(True)
save_fig("cnn_actual_vs_predicted_graph.png")


# ============================================================
# Random Forest Baseline Comparison
# ============================================================

X_ml = encoded_df.drop(columns=[target_col])
y_ml = encoded_df[target_col]

X_train_ml, X_test_ml, y_train_ml, y_test_ml = train_test_split(
    X_ml,
    y_ml,
    test_size=0.2,
    random_state=42,
    stratify=y_ml
)

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_ml, y_train_ml)

rf_pred = rf_model.predict(X_test_ml)
rf_accuracy = accuracy_score(y_test_ml, rf_pred)

apriori_score = validation_df["test_rule_accuracy"].mean() if not validation_df.empty else 0

comparison_df = pd.DataFrame({
    "Model": ["CNN", "Random Forest", "Apriori Rule Mining"],
    "Score": [cnn_accuracy, rf_accuracy, apriori_score]
})

save_csv(comparison_df, "model_comparison_results.csv")

plt.figure(figsize=(8, 5))
plt.bar(comparison_df["Model"], comparison_df["Score"])
plt.xlabel("Model")
plt.ylabel("Accuracy / Rule Score")
plt.title("CNN vs Random Forest vs Apriori")
save_fig("model_comparison_graph.png")


# ============================================================
# JSON and YAML Summary
# ============================================================

summary = {
    "project_title": "Global Pollution Analysis and Energy Recovery using Apriori Algorithm",
    "author": "Sagnik Patra",
    "dataset_used": "Global_Pollution_Analysis.csv",
    "raw_dataset_included": False,
    "readme_included": False,
    "zip_created": False,
    "cnn_accuracy": float(cnn_accuracy),
    "random_forest_accuracy": float(rf_accuracy),
    "apriori_average_rule_accuracy": float(apriori_score),
    "minimum_support": 0.05,
    "minimum_confidence": 0.40,
    "cnn_epochs": 30,
    "cnn_batch_size": 16
}

save_json(summary, "project_summary.json")
save_yaml(summary, "project_config.yaml")


# ============================================================
# Requirements File
# ============================================================

requirements = "\n".join([
    "pandas",
    "numpy",
    "matplotlib",
    "seaborn",
    "scikit-learn",
    "mlxtend",
    "tensorflow",
    "pyyaml",
    "nbformat"
])

requirements_path = os.path.join(OUTPUT_DIR, "requirements.txt")

with open(requirements_path, "w", encoding="utf-8") as file:
    file.write(requirements)

print("Saved:", requirements_path)


# ============================================================
# Final Report - Safe Version
# ============================================================

top_rules_text = rules.head(10).to_string(index=False) if not rules.empty else "No strong rules found."

report_lines = []

report_lines.append("# Final Report")
report_lines.append("")
report_lines.append("## Project Title")
report_lines.append("")
report_lines.append("Global Pollution Analysis and Energy Recovery using Apriori Algorithm")
report_lines.append("")
report_lines.append("## Author")
report_lines.append("")
report_lines.append("Sagnik Patra")
report_lines.append("")
report_lines.append("## Objective")
report_lines.append("")
report_lines.append("The goal of this project is to analyze global pollution data to understand the relationship between pollution levels such as air, water, and soil pollution and energy recovery across different countries.")
report_lines.append("")
report_lines.append("The Apriori Algorithm is used to uncover hidden associations between pollution factors, energy consumption, and energy recovery patterns. A CNN model is also implemented to classify pollution severity using engineered image-like features.")
report_lines.append("")
report_lines.append("## Dataset")
report_lines.append("")
report_lines.append("Dataset used: Global_Pollution_Analysis.csv")
report_lines.append("")
report_lines.append("The raw dataset is not included in the final submission package as per assignment recommendations.")
report_lines.append("")
report_lines.append("## Phase 1: Data Preprocessing and Feature Engineering")
report_lines.append("")
report_lines.append("The dataset was loaded and cleaned. Missing numerical values were handled using mean imputation, and categorical missing values were handled using mode imputation.")
report_lines.append("")
report_lines.append("Preprocessing steps:")
report_lines.append("")
report_lines.append("- Missing value handling")
report_lines.append("- Duplicate removal")
report_lines.append("- Categorical encoding")
report_lines.append("- Pollution index normalization")
report_lines.append("- Feature engineering")
report_lines.append("- Pollution severity categorization")
report_lines.append("")
report_lines.append("Created features:")
report_lines.append("")
report_lines.append("- Energy_Recovered_Per_Capita")
report_lines.append("- Energy_Consumption_Efficiency")
report_lines.append("- Average_Pollution_Index")
report_lines.append("- Air_Severity")
report_lines.append("- Water_Severity")
report_lines.append("- Soil_Severity")
report_lines.append("- Overall_Severity")
report_lines.append("- Energy_Recovery_Level")
report_lines.append("- Energy_Consumption_Level")
report_lines.append("")
report_lines.append("## Phase 2: Apriori Algorithm")
report_lines.append("")
report_lines.append("The Apriori Algorithm was applied to discover frequent itemsets and association rules.")
report_lines.append("")
report_lines.append("Minimum Support: 0.05")
report_lines.append("")
report_lines.append("Minimum Confidence: 0.40")
report_lines.append("")
report_lines.append("Association rule metrics:")
report_lines.append("")
report_lines.append("- Support")
report_lines.append("- Confidence")
report_lines.append("- Lift")
report_lines.append("- Leverage")
report_lines.append("- Conviction")
report_lines.append("")
report_lines.append("## Top Association Rules")
report_lines.append("")
report_lines.append("```text")
report_lines.append(top_rules_text)
report_lines.append("```")
report_lines.append("")
report_lines.append("## Phase 3: Model Evaluation and Validation")
report_lines.append("")
report_lines.append("The Apriori rules were validated using train-test split and 5-fold cross-validation.")
report_lines.append("")
report_lines.append("Apriori Average Rule Accuracy:")
report_lines.append("")
report_lines.append(str(apriori_score))
report_lines.append("")
report_lines.append("## Phase 4: CNN Classification")
report_lines.append("")
report_lines.append("A CNN model was trained by converting tabular pollution features into image-like matrices.")
report_lines.append("")
report_lines.append("CNN Accuracy:")
report_lines.append("")
report_lines.append(str(cnn_accuracy))
report_lines.append("")
report_lines.append("Random Forest Accuracy:")
report_lines.append("")
report_lines.append(str(rf_accuracy))
report_lines.append("")
report_lines.append("## Generated Visualizations")
report_lines.append("")
report_lines.append("- Pollution trend graph")
report_lines.append("- Correlation heatmap")
report_lines.append("- Frequent itemsets graph")
report_lines.append("- Association rules scatter graph")
report_lines.append("- Association rules lift graph")
report_lines.append("- Apriori cross-validation graph")
report_lines.append("- CNN confusion matrix")
report_lines.append("- CNN accuracy graph")
report_lines.append("- CNN loss graph")
report_lines.append("- CNN ROC curve")
report_lines.append("- CNN actual vs predicted graph")
report_lines.append("- Model comparison graph")
report_lines.append("")
report_lines.append("## Key Findings")
report_lines.append("")
report_lines.append("- Apriori provides explainable rules between pollution severity and energy recovery.")
report_lines.append("- CNN provides predictive classification of pollution severity.")
report_lines.append("- Random Forest provides a baseline comparison.")
report_lines.append("- Support, confidence, and lift help evaluate rule strength and usefulness.")
report_lines.append("- Rule-based learning and CNN-based prediction together provide stronger environmental analysis.")
report_lines.append("")
report_lines.append("## Actionable Recommendations")
report_lines.append("")
report_lines.append("- Prioritize high pollution regions with poor energy recovery.")
report_lines.append("- Improve renewable and waste-to-energy recovery systems.")
report_lines.append("- Use association rules to identify repeated pollution-energy patterns.")
report_lines.append("- Use CNN predictions for pollution severity monitoring.")
report_lines.append("- Apply cross-validation to ensure rule reliability before policy use.")
report_lines.append("")
report_lines.append("## Conclusion")
report_lines.append("")
report_lines.append("This project successfully applies Apriori Association Rule Mining, CNN classification, and validation techniques to analyze pollution and energy recovery relationships.")

final_report = "\n".join(report_lines)

report_path = os.path.join(OUTPUT_DIR, PREFIX + "final_report.md")

with open(report_path, "w", encoding="utf-8") as file:
    file.write(final_report)

print("Saved:", report_path)


# ============================================================
# Create Clean Jupyter Notebook Without Large Outputs
# ============================================================

nb = nbformat.v4.new_notebook()

nb.cells = [
    nbformat.v4.new_markdown_cell(
        "# Global Pollution Analysis and Energy Recovery using Apriori Algorithm\n\nAuthor: Sagnik Patra"
    ),
    nbformat.v4.new_markdown_cell(
        "## Objective\n\nAnalyze global pollution data using Apriori Algorithm and CNN classification."
    ),
    nbformat.v4.new_code_cell(
        "print('Notebook prepared for submission. Outputs are cleared to keep ZIP size small.')"
    ),
    nbformat.v4.new_markdown_cell(
        "## Generated Outputs\n\nGraphs, CSV metrics, model file, config files, and final report are generated by the main script."
    )
]

notebook_path = os.path.join(
    OUTPUT_DIR,
    "Global_Pollution_Apriori_CNN_Submission.ipynb"
)

with open(notebook_path, "w", encoding="utf-8") as file:
    nbformat.write(nb, file)

print("Saved clean notebook:", notebook_path)


# ============================================================
# Save Main Python Script Placeholder
# ============================================================

script_text = "\n".join([
    "# Global Pollution Analysis and Energy Recovery using Apriori Algorithm",
    "# Author: Sagnik Patra",
    "# This file is kept for submission support.",
    "# The main implementation is available in the executed notebook/cell.",
    "print('Global Pollution Apriori CNN project submission file.')"
])

script_path = os.path.join(OUTPUT_DIR, "pollution_apriori_cnn_submission.py")

with open(script_path, "w", encoding="utf-8") as file:
    file.write(script_text)

print("Saved:", script_path)


# ============================================================
# Submission File List
# ============================================================

submission_files = [
    "Global_Pollution_Apriori_CNN_Submission.ipynb",
    "pollution_apriori_cnn_submission.py",
    "requirements.txt",
    "apriori_final_report.md",
    "apriori_project_summary.json",
    "apriori_project_config.yaml",
    "apriori_processed_features.csv",
    "apriori_encoded_normalized_features.csv",
    "apriori_pollution_trend_summary.csv",
    "apriori_transaction_matrix.csv",
    "apriori_frequent_itemsets.csv",
    "apriori_association_rules.csv",
    "apriori_apriori_validation_results.csv",
    "apriori_apriori_cross_validation_results.csv",
    "apriori_cnn_classification_report.csv",
    "apriori_cnn_prediction_results.csv",
    "apriori_model_comparison_results.csv",
    "apriori_cnn_pollution_model.h5",
    "apriori_pollution_trend_graph.png",
    "apriori_correlation_heatmap.png",
    "apriori_frequent_itemsets_graph.png",
    "apriori_association_rules_scatter_graph.png",
    "apriori_association_rules_lift_graph.png",
    "apriori_apriori_cross_validation_graph.png",
    "apriori_cnn_confusion_matrix.png",
    "apriori_cnn_accuracy_graph.png",
    "apriori_cnn_loss_graph.png",
    "apriori_cnn_roc_curve.png",
    "apriori_cnn_actual_vs_predicted_graph.png",
    "apriori_model_comparison_graph.png"
]

submission_list_path = os.path.join(OUTPUT_DIR, "submission_file_list.txt")

with open(submission_list_path, "w", encoding="utf-8") as file:
    for item in submission_files:
        file.write(item + "\n")

print("Saved:", submission_list_path)


# ============================================================
# Excluded File List
# ============================================================

excluded_items = [
    "Global_Pollution_Analysis.csv",
    "README.md",
    "venv/",
    "env/",
    ".venv/",
    "node_modules/",
    "dist/",
    "build/",
    ".next/",
    ".cache/",
    "__pycache__/",
    ".ipynb_checkpoints/",
    "large media files",
    "raw data dumps"
]

excluded_list_path = os.path.join(OUTPUT_DIR, "excluded_file_list.txt")

with open(excluded_list_path, "w", encoding="utf-8") as file:
    for item in excluded_items:
        file.write(item + "\n")

print("Saved:", excluded_list_path)


# ============================================================
# Done
# ============================================================

print("\nDONE.")
print("Submit only files listed in submission_file_list.txt.")
print("Do NOT submit raw dataset, README, cache folders, virtual environment, or large media files.")
````
